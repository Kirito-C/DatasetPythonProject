What team has more win rounds and more lose rounds?

--------------------------------------------

Chi ha vinto e perso più partite?

```python
import requests
response = requests.get('https://raw.githubusercontent.com/Kirito-C/DatasetPythonProject/refs/heads/main/valorant_champions_istanbul.json')
data = response.json()

win_rounds = 0
win_team = ""
lose_rounds = 0
lose_team = ""

"Rounds Win"

for player in data:
  if player['Rounds Win'] > win_rounds:
    win_rounds = player['Rounds Win']
    win_team = player['Team']
  if player['Rounds Lose'] > lose_rounds:
    lose_rounds = player['Rounds Lose']
    lose_team = player['Team']

print(f"Team: {win_team} - wins rounds: {win_rounds}")
print(f"Team: {lose_team} - lose rounds: {lose_rounds}")
```

What is KAST?
Used also in CS:GO, KAST is a metric that calculates the contribution of each player per round based on the number of rounds they played. It is basically shorthand for KILLS, ASSISTS, SURVIVED and TRADED.

```python
import requests
response = requests.get('https://raw.githubusercontent.com/Kirito-C/DatasetPythonProject/refs/heads/main/valorant_champions_istanbul.json')
data = response.json()

def printMenu():
  print()
  print("1. print all player's name / stampa i nomi di tutti i giocatori")
  print("2. print all player's stats / stampa le statistiche di tutti i giocatori")
  print("3. print all infos of a player / stampa tutte le informazioni di un giocatore")
  print("4. print the stats of a player / stampa le statistiche di un giocatore")
  print("5. exit / esci")
  print()

while True:
    printMenu()
    choice = input("Select an option / Scegli un opzione: ")

    match choice:
        case "1":
            print()
            for player in data:
                print(f"- name: {player['Player']}")
                print()

        case "2":
            print()
            for player in data:
              print(f"name: {player['Player']}:")
              print(f"  - Kills: {player['Kill']}")
              print(f"  - Deaths: {player['Death']}")
              print(f"  - K|D: {player['K|D']}")
              print(f"  - KAST: {player['KAST']}")
              print(f"  - Role: {player['Role']}")
              print(f"  - Headshot percentages (HS%): {player['HS %']}")
              print(f"  - Rounds played: {player['Rounds Played']}")
              print(f"  - Rounds Win: {player['Rounds Win']}")
              print(f"  - Rounds Lose: {player['Rounds Lose']}")
              print(f"  - Rank: {player['Rank']}")
              print()

        case "3":
            print()
            player_name = input("Select a player / Seleziona un giocatore:")
            for player in data:
              if player['Player'] == player_name:
                print(f"Name: {player['Player']}")
                print(f"- Nationality: {player['Nationality']}")
                print(f"- Team: {player['Team']}")
                print(f"- Rank: {player['Rank']}")
                print()
            print()

        case "4":
            print()
            player_name = input("Select a player / Seleziona un giocatore:")
            found = False
            for player in data:
                if player['Player'] == player_name:
                  found = True
                  print(f"name: {player['Player']}:")
                  print(f"  - Kills: {player['Kill']}")
                  print(f"  - Deaths: {player['Death']}")
                  print(f"  - K|D: {player['K|D']}")
                  print(f"  - KAST: {player['KAST']}")
                  print(f"  - Role: {player['Role']}")
                  print(f"  - Headshot percentages (HS%): {player['HS %']}")
                  print(f"  - Rounds played: {player['Rounds Played']}")
                  print(f"  - Rounds Win: {player['Rounds Win']}")
                  print(f"  - Rounds Lose: {player['Rounds Lose']}")
                  print(f"  - Rank: {player['Rank']}")

            if found == False:
              print()
              print("No Player found / Nessun giocatore trovato")
            print()

        case "5":
          break

        case _:
          print("You haven't chose any options / Non hai scelto alcuna opzione")
```

Map that contains some points, each country a player.

-------------------------------------------------

Mappa contenente punti su ogni Nazione per un giocatore.

```python
import folium
import requests

request = requests.get('https://raw.githubusercontent.com/Kirito-C/DatasetPythonProject/refs/heads/main/valorant_champions_istanbul.json')
data_players = request.json()

request = requests.get('https://raw.githubusercontent.com/Kirito-C/DatasetPythonProject/refs/heads/main/countries_with_coordinates.json')
data_countries = request.json()

teams = set()

for player in data_players:
  teams.add(player['Team'])

teams = list(teams)
countries = []
players = []

colors = ['orange', 'darkgreen', 'red', 'blue', 'pink', 'darkblue', 'purple', 'black']

for team in teams:
  for player in data_players:
    if player['Team'] == team:
      players.append(player)
      for country in data_countries:
        if player['Nationality'] == country['country']:
          countries.append(country)
          break
      break

map = folium.Map(
    control_scale=True
)

for player in players:
  for country in countries:
    if player['Nationality'] == country['country']:
      folium.Marker(
        location=[country['latitude'], country['longitude']],
        tooltip=player['Team'],
        popup=player['Player'] + " Rank: " + player['Rank'],
        max_width=200,
        icon=folium.Icon(color=colors[0]),
      ).add_to(map)
      countries.pop(countries.index(country))
      colors.pop(0)
      break

map
```

The users can choose three players and see different graphs of a three topics chosen by the user himself

--------------------------------------------------

l'utente può scegliere tre giocatori e vedere grafici diversi di tre argomenti scelti dall'utente stesso


```python
import requests
import matplotlib.pyplot as plt
import numpy as np

response = requests.get('https://raw.githubusercontent.com/Kirito-C/DatasetPythonProject/refs/heads/main/valorant_champions_istanbul.json')
data = response.json()

def printMenu():
  print()
  print("1. print all player's name / stampa i nomi di tutti i giocatori")
  print("2. change the three players chosen / cambia i tre giocatori scelti")
  print("3. change the topics chosen before / cambia gli argomenti scelti prima")
  print("4. see the three players chosen / vedi i tre ragazzi scelti")
  print("5. see the chosen topics / vedi gli argomenti scelti")
  print("be aware that at the start of the program the three topics and the three players must have been chosen to see the graphs")
  print("tenere presente che all'inizio del programma devono essere stati scelti i tre argomenti e i tre giocatori per poter vedere i grafici")
  print("6. print the graphs / stampa i grafici")
  print("7. exit / esci")
  print()

def printTopics():
  print()
  for topic in topics:
    print(f"- {topic}")
    print()
  print()

three_players = []
topics = ['Kill', 'Death', 'K|D', 'HS %', 'Rounds Played', 'Rounds Win', 'Rounds Lose', 'Rank']
three_topics = []

while True:
    printMenu()
    choice = input("Select an option / Scegli un opzione: ")

    match choice:
        case "1":
            print()
            for player in data:
                print(f"- name: {player['Player']} - Team: {player['Team']}")
                print()
            print()

        case "2":
            print()
            if len(three_players) == 3:
              three_players = []

            for i in range(3):
              found = False
              while True:
                print()
                player_name = input("Select a player / Seleziona un giocatore:")
                for player in data:
                  if player['Player'] == player_name:
                    found = True
                    three_players.append(player)
                    break
                if found:
                  break
                else:
                  print()
                  print("No player found / Nessun giocatore trovato")
            print()
            print(f"Players inserted / giocatori inseriti: {three_players[0]['Player']} - {three_players[1]['Player']} - {three_players[2]['Player']}")

        case "3":
             print()
             if len(three_topics) == 3:
              three_topics = []

             printTopics()

             for i in range(3):
              found = False

              while True:
                topic = input("Select a topic / Seleziona un argomento:")
                if topic in topics:
                  found = True
                  three_topics.append(topic)
                  break
                else:
                  print("No topic found / Nessun argomento trovato")

             print()
             print(f"Topics inserted / argomenti inseriti: {three_topics[0]} - {three_topics[1]} - {three_topics[2]}")

        case "4":
            print()
            print(f"Players chosen / Giocatori scelti: {three_players[0]['Player']} - {three_players[1]['Player']} - {three_players[2]['Player']}")
            print()

        case "5":
            print()
            print(f"Topics chosen / Argomento scelti: {three_topics[0]} - {three_topics[1]} - {three_topics[2]}")
            print()

        case "6":
            if not three_players:
              print()
              print("No players to show on graphs (list empty) / Nessun giocatore da mostrare sui grafici (lista vuota)")
              print()
            else:
              three_players_names = []
              for player in three_players:
                three_players_names.append(player['Player'])

              fig, (ax1, ax2, ax3) = plt.subplots(nrows=1, ncols=3, sharex=True,
                                    figsize=(15, 15))

              # first graph / primo grafico

              bar_labels = ['red', 'blue', 'orange']
              bar_colors = ['tab:red', 'tab:blue', 'tab:orange']

              first_player = float(three_players[0][three_topics[0]])
              second_player = float(three_players[1][three_topics[0]])
              third_player = float(three_players[2][three_topics[0]])

              first_topic_players = [first_player, second_player, third_player]


              ax1.bar(three_players_names, first_topic_players, label=bar_labels, color=bar_colors)

              ax1.set_ylabel(three_topics[0])
              ax1.set_title(three_topics[0])
              ax1.legend(title='Players')

              # second graph / secondo grafico

              first_player = float(three_players[0][three_topics[1]])
              second_player = float(three_players[1][three_topics[1]])
              third_player = float(three_players[2][three_topics[1]])

              second_topic_players = [first_player, second_player, third_player]

              ax2.bar(three_players_names, second_topic_players, label=bar_labels, color=bar_colors)

              ax2.set_ylabel(three_topics[1])
              ax2.set_title(three_topics[1])
              ax2.legend(title='Players')

              # third graph / terzo grafico
              #controllo perché se non cerco io tramite la chiave in formato stringa il programma mi ritorna errore dicendomi che
              #il terzo topic come chiave vale 'K\\/D' stranamente. Ho provato a stampare three_topics[2] e mi stampa normalmente
              #'K\/D' ma quando la uso come chiave cambia

              first_player = float(three_players[0][three_topics[2]])
              second_player = float(three_players[1][three_topics[2]])
              third_player = float(three_players[2][three_topics[2]])

              third_topic_players = [first_player, second_player, third_player]


              ax3.bar(three_players_names, third_topic_players, label=bar_labels, color=bar_colors)

              ax3.set_ylabel(three_topics[2])
              ax3.set_title(three_topics[2])
              ax3.legend(title='Players')

              plt.show()




        case "7":
            break

        case _:
          print("You haven't chose any options / Non hai scelto alcuna opzione")
```
